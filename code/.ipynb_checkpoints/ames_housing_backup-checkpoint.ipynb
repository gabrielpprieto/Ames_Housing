```python
# Converto 'Numerical' columns - Into Strings
def column_to_type(dataframe, col_type, columns):
    for column in columns:
        dataframe[column] = dataframe[column].apply(col_type)

column_to_type(train, str, ['mssubclass', 'yrsold', 'mosold'])
column_to_type(test, str, ['mssubclass', 'yrsold', 'mosold'])
```

```python
# define numerical and nominal columns
numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerical_variables = train.select_dtypes(include=numeric_dtypes).columns
nominal_variables = train.select_dtypes(exclude=numeric_dtypes).columns
```

```python
# # create columns to show which columns had missing values
# def missing_values(dataframe, null_columns):
#     for column in null_columns:
#         dataframe[column + '_was_missing'] = dataframe[column].isnull().astype(int)
        

# missing_values(train, null_train)
# missing_values(test, null_test)
```

```python
# impute missing values
def input_missing_values(dataframe):
    
    # Adapted From: https://www.kaggle.com/lavanyashukla01/how-i-made-top-0-3-on-a-kaggle-competition
    # Group the by neighborhoods, and fill in missing value by the median LotFrontage of the neighborhood
    dataframe['lotfrontage'] = dataframe.groupby('neighborhood')['lotfrontage'] \
    .transform(lambda x: x.fillna(x.median()))
    
    dataframe['mszoning'] = dataframe.groupby('mssubclass')['mszoning'] \
    .transform(lambda x: x.fillna(x.mode()[0]))
    
    dataframe['masvnrarea'] = dataframe['masvnrarea'].fillna(0)
    
    dataframe['electrical'] = dataframe['electrical'].fillna(lambda x: x.fillna(x.mode))
    
    for col in ['garageyrblt', 'garagearea', 'garagecars']:
        dataframe[col] = dataframe[col].fillna(0)

    # Input Missing columns according to being Numeric or Nominal
    for column in dataframe.isnull().mean()[dataframe.isnull().mean() > 0].sort_values(ascending=False).index:
        if column in numerical_variables:
            dataframe[column] = dataframe[column].fillna(0)
        elif column in nominal_variables:
            dataframe[column] = dataframe[column].fillna('None')

input_missing_values(train)
input_missing_values(test)
```

```python
def new_features(*args):
    for arg in args:
        arg['totalarea'] = arg['totalbsmtsf'] + arg['1stflrsf'] + arg['2ndflrsf']
                
        arg['remodminusbuilt'] = arg['yearremodadd'] - arg['yearbuilt']

        arg['totalsf'] = (arg['bsmtfinsf1'] + arg['bsmtfinsf2'] + arg['1stflrsf'] + arg['2ndflrsf'])

        arg['totalbathrooms'] = (arg['fullbath'] + (0.5 * arg['halfbath']) + arg['bsmtfullbath'] +
                                  (0.5 * arg['bsmthalfbath']))

        arg['totalporchsf'] = (arg['openporchsf'] + arg['3ssnporch'] + arg['enclosedporch'] +
                                 arg['screenporch'] + arg['wooddecksf'])
        
        arg['haspool'] = arg['poolarea'].apply(lambda x: 1 if x > 0 else 0)
        arg['has2ndfloor'] = arg['2ndflrsf'].apply(lambda x: 1 if x > 0 else 0)
        arg['hasgarage'] = arg['garagearea'].apply(lambda x: 1 if x > 0 else 0)
        arg['hasbsmt'] = arg['totalbsmtsf'].apply(lambda x: 1 if x > 0 else 0)
        arg['hasfireplace'] = arg['fireplaces'].apply(lambda x: 1 if x > 0 else 0)

new_features(poly_train, poly_test)
```

#### Modeling

```python
# Define X and y
X = poly_train.drop('saleprice', axis=1)
y = train['saleprice']

# Model Prep: Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.20, shuffle=True)

# transforming y to y_log - normalizing distribution
y_train_log = y_train.map(np.log1p)
y_test_log = y_test.map(np.log1p)
y_log = np.log1p(y)
```

```python
# Null model - Baseline
y_train.mean()
```

#### GridSearchCV - Set Up Function to Run Models

```python
# Define function to run Pipeline + GridSearchCV
def run_model_gs(pipeline, parameters, cv=7, X_train=X_train, X_test=X_test, y_train=y_train_log, y_test=y_test_log):
    
    # Instantiate RandomizedSearchCV
    gs = GridSearchCV(pipeline,
                      parameters,
                      cv=cv,
                      verbose=1,
                      n_jobs=-1)
    
    # Fit GridSearch to training data
    gs.fit(pd.DataFrame(X_train, columns=X_train.columns), pd.Series(y_train_log, name='saleprice'))
    
           
    # Best mean score out of cross validation out of all models tested (cvec)
    print(f'Best Score: {round(gs.best_score_,4)}')
    
    # Best Paramenters
    print(f'Best Parameters: {gs.best_params_}')
    
    # Save best model as gs_cvec_logreg_model
    gs_model = gs.best_estimator_
    
    # Score model on training set.
    print(f'Training Accuracy Score: {round(gs_model.score(X_train, y_train_log),4)}')

    # Score model on testing set.
    print(f'Testing Accuracy Score: {round(gs_model.score(X_test, y_test_log),4)}')
    
    # Predictions
    preds = gs_model.predict(X_test)
    
    print(f'Testing RMSE: {round(np.sqrt(mean_squared_error(y_test_log, preds)),4)}')
    print(f'Testing RMSLE: {round(np.sqrt(mean_squared_log_error(y_test_log, preds)),4)}')

    return gs_model
```

```python
pipeline_ridge = Pipeline(steps=[('scaler', RobustScaler()),
                                 ('lr', Ridge())])

params_ridge = {
    'lr__alpha': [x for x in range(1,21)],
}
```

```python
ridge = run_model_gs(pipeline_ridge, params_ridge)
```

```python
pipeline_lasso = Pipeline(steps=[('scaler', RobustScaler()),
                                 ('lr', Lasso())])

params_lasso = {
    'lr__alpha': [0.0008,0.00095,0.001,0.01,0.1,1,10,100],
}
```

```python
lasso = run_model_gs(pipeline_lasso, params_lasso)
```

```python
pipeline_enet = Pipeline(steps=[('scaler', RobustScaler()),
                                ('enet', ElasticNet())])

params_enet = {
    'enet__alpha': [0.0001, 0.0002, 0.0003, 0.0004, 0.0005, 0.0006, 0.0007],
    'enet__l1_ratio': [0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 0.99, 1]
}
```

```python
enet = run_model_gs(pipeline_enet, params_enet)
```

```python
pipeline_svr = Pipeline(steps=[('scaler', RobustScaler()),
                                ('svr', SVR())])

params_svr = {
    'svr__kernel': ['linear'],
    'svr__gamma': ['scale'],
    'svr__C': [0.1],
    'svr__epsilon': [0.1]
}
```

```python
svr = run_model_gs(pipeline_svr, params_svr)
```

```python
pipeline_xgb = Pipeline(steps=[('scaler', RobustScaler()),
                               ('xgb', XGBRegressor())])

params_xgb = {
    'xgb__max_depth': [4],
    'xgb__learning_rate': [0.05, 0.08],
    'xgb__n_estimators': [200],
    'xgb__colsample_bytree': [1],
    'xgb__alpha': [0.1],
    'xgb__min_child_weight': [1],
    'xgb__gamma': [0],
    'xgb__subsample': [1],
    'xgb__colsample_bytree': [1]
    }
```

```python
xgb = run_model_gs(pipeline_xgb, params_xgb)
```

```python
pipeline_lgbm = Pipeline(steps=[('scaler', RobustScaler()),
                                ('lgbm', LGBMRegressor())])

params_lgbm = {
   'lgbm__objective': ['regression'], 
   'lgbm__num_leaves' : [4, 5],
   'lgbm__learning_rate' : [0.01, 0.05], 
   'lgbm__n_estimators': [1000],
   'lgbm__max_bin' : [200], 
}
```

```python
lgbm = run_model_gs(pipeline_lgbm, params_lgbm)
```

#### Voting Regressor - Ensembing

```python
vote = VotingRegressor([('Ridge', ridge), ('Lasso', lasso), ('ElasticNet', enet),
                        ('SVR', svr),('XGBRegressor', xgb), ('LGBMRegressor', lgbm)],
                        n_jobs=-1)

vote = vote.fit(X_train, y_train_log)

# Score model on training set.
print(f'Training Accuracy Score: {vote.score(X_train, y_train_log)}')

# Score model on testing set.
print(f'Testing Accuracy Score: {vote.score(X_test, y_test_log)}')

# Predictions
preds = vote.predict(X_test)

print(f'RMSE: {np.sqrt(mean_squared_error(y_test_log, preds))}')
print(f'RMSLE: {np.sqrt(mean_squared_log_error(y_test_log, preds))}')
```
